##4) Preprocessing Steps (10 pts)

As some of the features (predictive variables) on this dataset are categorical (non-numeric) you need to do some preprocessing for those features.

You can use as many **dummy or indicator variables** as there are categories within one feature. You can also look at pandas' get_dummies or keras.utils.to_categorical functions.

In neural networks, scaling of the features are important, because they affect the net input of a neuron as a whole. You should use **MinMax scaler** on sklearn for this task, which scales the variables between 0 and 1 on by default. (Remember that mean-squared error loss function tends to be extremely large with unscaled features.)


```python
# encode the categorical variables

# scale the features between 0-1
from sklearn.preprocessing import MinMaxScaler
msc = MinMaxScaler() 

print("Scale the features between 0-1:","\n",)
train_df1 = pd.get_dummies(train_df, columns = ['crime_rate'])
train_df2 = pd.get_dummies(train_df1, columns = ['view'])
train_df2

```

```python
train_df2[["sqmtrs", "nrooms","price"]] = msc.fit_transform(train_df2[["sqmtrs", "nrooms","price"]])
train_df2
```

Don't forget the split the training data to obtain a validation set. **Use random_state=42**

```python
# split 90-10

# Define X:
X = train_df2[['sqmtrs', 'nrooms']]
# Define y:
y = train_df2['price']

from sklearn.model_selection import train_test_split
# Split 90-10
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

##5) Train neural networks on development data and do model selection using the validation data (55 pts)


* Train a neural network with **one hidden layer** (try 3 different values for the number of neurons in that hidden layer, as 25, 50, 100), you will need to correctly choose the optimizer and the loss function that this model will train with. Use batch_size as 64 and train each model for 30 epochs. 

* Train another neural network with two hidden layers with meta-parameters of your choice. Again, use batch_size as 64 and train the model for 30 epochs. 

* **Bonus (5 pts)** Train a KNN or a Decision Tree model with your own choice of meta parameters to predict the house prices.


```python
# define your model architecture
import keras
import keras.utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split
from keras.datasets import mnist
from sklearn.utils import shuffle

histories=[]

model1 = Sequential()
model1.add(Flatten())

# train one-hidden layered neural networks
model1.add(Dense(25, activation='relu', input_shape=(X_train.shape[1],), name='hidden_1'))
```

```python
# compile your model with an optimizer
model1.compile(loss='mean_squared_error', optimizer="adam")
```

```python
# fit the model on training data
history1 = model1.fit(X_train, y_train, batch_size=64, epochs=30, validation_split = 0.1, verbose=1)
histories.append(history1)
```

```python
score1 = model1.evaluate(X_test, y_test, verbose=0)
print('Validation loss:', score1)
```

```python
model2 = Sequential()
model2.add(Flatten())
# train a two-hidden layered neural network
model2.add(Dense(50, activation='relu', input_shape=(X_train.shape[1],), name='hidden_2'))

# compile your model with an optimizer
model2.compile(loss='mean_squared_error', optimizer="adam")

# fit the model on training data
history2 = model2.fit(X_train, y_train, batch_size=64, epochs=30, validation_split = 0.1, verbose=1)
histories.append(history2)
```

```python
score2 = model2.evaluate(X_test, y_test, verbose=0)
print('Validation loss:', score2)
```

```python
model3 = Sequential()
model3.add(Flatten())
# train a three-hidden layered neural network
model3.add(Dense(100, activation='relu', input_shape=(X_train.shape[1],), name='hidden_3'))

# compile your model with an optimizer
model3.compile(loss='mean_squared_error', optimizer="adam")

# fit the model on training data
history3 = model3.fit(X_train, y_train, batch_size=64, epochs=30, validation_split = 0.1, verbose=1)
histories.append(history3)
```

```python
score3 = model3.evaluate(X_test, y_test, verbose=0)
print('Validation loss:', score3)
```

```python
model4 = Sequential()
model4.add(Flatten())
model4.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],), name='hidden_1'))
model4.add(Dense(128, activation='relu', name='hidden_2'))
model4.add(Dense(1, name='output_layer'))
# compile your model with an optimizer
model4.compile(loss='mean_squared_error', optimizer="adam")
# fit the model on training data
history4 = model4.fit(X_train, y_train, batch_size=64, epochs=30, validation_split = 0.1, verbose=1)
histories.append(history4)
```

```python
score4 = model4.evaluate(X_test, y_test, verbose=0)
print('Validation loss:', score4)
```

## 6) Test your trained classifiers on the Validation set (10 pts)
Test your trained classifiers on the validation set and print the mean squared errors.


```python
# tests on validation
y_pred1 = model1.predict(X_test).sum(axis=1)
y_pred2 = model2.predict(X_test).sum(axis=1)
y_pred3 = model3.predict(X_test).sum(axis=1)
y_pred4 = model4.predict(X_test).sum(axis=1)

from sklearn import metrics
print('MSE is :', metrics.mean_squared_error(y_test, y_pred1)) 

print('MSE is :', metrics.mean_squared_error(y_test, y_pred2)) 

print('MSE is :', metrics.mean_squared_error(y_test, y_pred3)) 

print('MSE is :', metrics.mean_squared_error(y_test, y_pred4)) 
```
